 <h2> OpenCV </h2>

```python
import cv2
```

```python
img = cv2.imread('/content/fruit.jpg')
```

```python
type(img)
```

```python
img.shape
```

**Resizing the image**

```python
img_resized = cv2.resize(img, (200,200) )
```

```python
img_resized.shape
```

 **Converting RGB images to Grayscale image using OpenCV**

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image
cv2.imwrite('fruit_grayscale_image.jpg', grayscale_image)
```
